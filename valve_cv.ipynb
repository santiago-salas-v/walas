# Unit conversion for the valve coefficient

EB &emsp; $0~m=\frac{\Delta p}{\rho g}+\Delta z+\frac{\Delta w^2}{2 g}+\Delta H_{v}+\frac{\dot Q}{\rho g \dot V}+\frac{C_v\Delta T}{g}+\frac{-\dot W_{Welle}}{\rho g \dot V}$

Anlagenkennlinie $H_{Anlage}=\underbrace{\Delta z+\frac{p_{aus}-p_{ein}}{\rho g}}_{\text{stat. Förderhöhe, }f(\Delta z)}+\underbrace{\frac{w_{aus}^2-w_{ein}^2}{2g}+\sum H_{Verluste}}_{\text{dyn. Förderhöhe, }f(\dot V)}$

Pumpenkennlinie $H_{Pumpe}=H(\dot V)$ 

p is pressure, $\rho$ is density, g is gravitational acceleration, z is elevation, w is linear velocity, $H_{Verluste}$ are head losses due to friction, $\dot Q$ is heat flow into the system, $\dot V$ is volume flow, $C_v\Delta T$ is change in internal energy due solely to temperature change of incompressible fluid, $\dot W_{Welle}$ is power given by shaft work. $H$ are pressure heads as heights. 

Pump curve gives a decreasing dynamic head $H_{Pumpe}$ with increasing volume flow $\dot V$. Plant curve $H_{Anlage}$ consists of a static head dependent on plant conditions like height, inlet and outlet pressure (independent of flow), and a dynamic head incresing with volume flow $\dot V$. Pump must be able to meet the dynamic + static head of the plant at a given flow.

Note: Below, as in ref. [1] volume flow is $Q$. 

Definition of valve flow coefficients, according to [1].

* Resistance coefficient $\zeta$

$\Delta h=\zeta \frac{v^2}{2g}$

* Flow coefficient Cv

$C_v=Q\sqrt{\frac{\Delta p_0}{\Delta p}\frac{\rho}{\rho_0}}$

Q &emsp;&emsp; flow in US gal/min

$\Delta p_0$ &emsp; reference differential pressure = 1lb/in²

$\Delta p_0$ &emsp; operating differential pressure in lb/in²

$\rho_0$ &emsp; &emsp; density of reference fluid water = 62.4 lg/ft³

$\rho$ &emsp; &emsp; density of operating fluid in lb/ft³
 

* Flow coefficient Kv

$K_v=Q\sqrt{\frac{\Delta p_0}{\Delta p}\frac{\rho}{\rho_0}}$

Q &emsp;&emsp; flow in m³/h

$\Delta p_0$ &emsp; reference differential pressure = 1bar

$\Delta p_0$ &emsp; operating differential pressure in bar

$\rho_0$ &emsp; &emsp; density of reference fluid water = 1000 kg/m³

$\rho$ &emsp; &emsp; density of operating fluid in kg/m³

* Flow coefficient Av

$A_v=Q\sqrt{\frac{\rho}{\Delta p}}$

Q &emsp;&emsp; flow in US gal/min

$\Delta p_0$ &emsp; reference differential pressure = 1lb/in²

$\Delta p_0$ &emsp; operating differential pressure in lb/in²

$\rho_0$ &emsp; &emsp; density of reference fluid water = 62.4 lg/ft³

$\rho$ &emsp; &emsp; density of operating fluid in lb/ft³
