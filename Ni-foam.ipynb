# Conductivity in metal foams by series-parallel approach

The following calculaions are based on publications [1] to [7]

## Properties of gas mixture at 1000K

### Viscosity 

Pure monatomic gas [1] with mass of a single molecule $m=M/N_A$ where $M$ is the molar mass and $N_A$ the Avogadro constant.

$\begin{align}
\mu=&\frac{5}{16}\frac{\sqrt{\pi m \kappa T}}{\pi\sigma^2\Omega_\mu}\\
=&\frac{5}{16}\sqrt{\frac{1.3806\cdot 10^{-23}}{\pi\cdot 1000\cdot 6.022\cdot 10^{23}}}\frac{1}{(10^{-10})^2}\frac{kg}{m\cdot s}\cdot\frac{\sqrt{\left(\frac{M}{g/mol}\right)\left(\frac{T}{K}\right)}}{\left(\frac{\sigma}{\unicode{x212B}}\right)^2\Omega_\mu}\\
=&\underset{2.6696\cdot 10^{-5}}{\underbrace{\frac{5}{16}\sqrt{\frac{1.3806\cdot 10^{-23}}{\pi\cdot 1000\cdot 6.022\cdot 10^{23}}}\frac{1}{(10^{-10})^2}\frac{1000}{100}}}\frac{g}{cm\cdot s}\cdot\frac{\sqrt{\left(\frac{M}{g/mol}\right)\left(\frac{T}{K}\right)}}{\left(\frac{\sigma}{\unicode{x212B}}\right)^2\Omega_\mu}
\end{align}$

with $\kappa$ Boltzmann constant, $\sigma$ Lennard-Jones collision diameter, $\epsilon$ Lennard-Jones maximum energy of atraction between a pair of molecules, $\Omega_\mu$ collision integral for viscosity.

Empirical equation for the collision integral for viscosity [2]:

$\Omega_\mu=A\cdot(T^*)^{-B}+C\cdot exp(-D\cdot(T^*))+E\cdot exp(-F\cdot(T^*))$

$(T^*)=\kappa T/\epsilon, A=1.16145, B=0.14874, C=0.52487, D=0.77320, E=2.16178, and F=2.43787$ applicable for $0.3\le(T^*)\le 100$.

Viscosity of the gas mixture with molar fractions $x_i$ at low density by the semiempirical relation [1]:

$\mu_{mix}=\sum_{\alpha=1}^{N}\frac{x_\alpha \mu_\alpha}{\sum_\beta x_\beta\Phi_{\alpha\beta}}$

$\Phi_{\alpha\beta}=\frac{1}{\sqrt{8}}\left(1+\frac{M_\alpha}{M_\beta}\right)^{-1/2}\left[1+\left(\frac{\mu_\alpha}{\mu_\beta}\right)^{1/2}\left(\frac{M_\beta}{M_\alpha}\right)^{1/4}\right]^2$

Lennard-Jones parameters [2]

i | $\sigma/\unicode{x212B}$ | $(\frac{\epsilon}{\kappa}) / K$
- | - | -
$H_2$ | 2.827 | 59.7
$CH_4$ | 3.758 | 148.6
$CO_2$ | 3.941 | 195.2
$CO$ | 3.690 | 91.7
$H_2O$ | 2.641 | 809.1
$N_2$ | 3.798 | 71.4
$O_2$ | 3.467 | 106.7

### Density

Ideal gas.

$\rho=\frac{M}{V}=\frac{p}{RT}M$

Density of mixture with molar fractions $x_i$:

$\rho=\frac{p}{RT}\sum_i x_i M_i$

### Heat capacity (constant pressure)

"NASA-polynomials" with coefficients from [3], applicable between 200K and 1000K, and between 1000K and 6000K.

$\frac{Cp}{R}=a_1+a_2 T+a_3 T^2+a_4 T^3+a_5 T^4$

Heat capacity of the gas mixture with molar fractions $x_i$:

$Cp_{mix}=\sum_i x_i Cp_i$

### Thermal conductivity

Three correlations to test, based on Eucken's semiempirical relation for polyatomic gases at low density, see ref. [1] and [2]. Using eq. 10-3.2 in [2] as basis:

$\frac{\lambda M}{\eta C_v}=\frac{15/4}{C_v/R}+f_{int}\left(1-\frac{3/2}{C_v/R}\right)$

Here, $M$ is the molar mass, $\mu$ is the dynamic viscosity, $C_v$ is the constant volume heat capacity, $\frac{\lambda M}{\eta C_v}$ is the Eucken number and $f_{int}$ represents the contribution due to internal degrees of freedom. The contribution due to translational degrees of freedom, $f_{tr}$, is set to $f_{tr}=2.5$, because the limiting form for monatomic gases at low pressure is $\frac{\lambda M}{\eta C_v}=2.5$. Additionally, the relationship between $C_v$ and $C_p$ for an ideal gas is: $C_p=C_v-T \left(\frac{\partial V}{\partial T}\right)^2_P/\left(\frac{\partial V}{\partial P}\right)_T=C_v+R$.

0. Monatomic gas: $C_v=\frac{3}{2}R$ and $f_{int}=0$

$\frac{\lambda M}{\eta C_v}=\frac{15/4}{3/2R}R=\frac{5}{2}$

1. Polyatomic gas by Eucken correlation: $f_{int}=1$

$\begin{align}
\lambda&=\left(\frac{15}{4}R+C_v-\frac{3}{2}R+\right)\frac{\mu}{M}=\left(\frac{9}{4}R+C_p-R\right)\frac{\mu}{M}\\
&=\left(Cp+\frac{5}{4}R\right)\mu\\
&=\left(\frac{Cp}{J/mol/K}+\frac{5}{4}\frac{R}{J/mol/K}\right)\frac{\mu}{kg/m/s}\frac{g/mol}{M}\cdot \frac{J}{mol K}\frac{kg}{m s}
\frac{mol}{g}\frac{1000g}{1kg}[=]\frac{W}{m K}\end{align}$

2. Polyatomic gas by modified Eucken correlation: $f_{int}\approx 1.32$

$\begin{align}
\lambda&=\left(1.32 C_p+\left[\frac{15}{4}-1.32-1.32\cdot\frac{3}{2}\right]R\right)\frac{\mu}{M}\\
&=\left(1.32\frac{Cp}{J/mol/K}+0.45\frac{R}{J/mol/K}\right)\frac{\mu}{kg/m/s}\frac{g/mol}{M}\cdot \frac{J}{mol K}\frac{kg}{m s}
\frac{mol}{g}\frac{1000g}{1kg}[=]\frac{W}{m K}\end{align}$

3. Polyatomic non-polar gas compromise (Stiel, Thodos [2]): $f_{int}\approx 1.15$

$\begin{align}
\lambda&=\left(1.15 C_p+\left[\frac{15}{4}-1.15-1.15\cdot\frac{3}{2}\right]R\right)\frac{\mu}{M}\\
&=\left(1.15\frac{Cp}{J/mol/K}+0.875\frac{R}{J/mol/K}\right)\frac{\mu}{kg/m/s}\frac{g/mol}{M}\cdot \frac{J}{mol K}\frac{kg}{m s}
\frac{mol}{g}\frac{1000g}{1kg}[=]\frac{W}{m K}\end{align}$

Conductivity of the gas mixture with molar fractions $x_i$ at low density by the semiempirical relation[1]:

$\lambda_{mix}=\sum_{\alpha=1}^{N}\frac{x_\alpha \lambda_\alpha}{\sum_\beta x_\beta\Phi_{\alpha\beta}}$

$\Phi_{\alpha\beta}=\frac{1}{\sqrt{8}}\left(1+\frac{M_\alpha}{M_\beta}\right)^{-1/2}\left[1+\left(\frac{\mu_\alpha}{\mu_\beta}\right)^{1/2}\left(\frac{M_\beta}{M_\alpha}\right)^{1/4}\right]^2$

Example with two components

$\lambda_{mix}=\frac{x_1\lambda_1}{x_1+x_2\frac{1}{\sqrt{8}}\left(1+\frac{M_1}{M_2}\right)^{-1/2}\left[1+\left(\frac{\mu_1}{\mu_2}\right)^{1/2}\left(\frac{M_2}{M_1}\right)^{1/4}\right]^2}+\frac{x_2\lambda_2}{x_1\frac{1}{\sqrt{8}}\left(1+\frac{M_2}{M_1}\right)^{-1/2}\left[1+\left(\frac{\mu_2}{\mu_1}\right)^{1/2}\left(\frac{M_1}{M_2}\right)^{1/4}\right]^2+
x_2}$

since $\frac{1}{\sqrt{8}}\left(1+1\right)^{-1/2}\left[1+1\right]^2=1$

## Effective heat conductivity of metal foam

Equivalence to VDI-Wärmeatlas nomenclature

$\lambda_p$: heat conductivity of particles

$\lambda_f$: heat conductivity of fluid filling void spaces

$\psi$: Porosity of the bed

The analytical solution by Maxwell (1873) is valid for diluted suspensions or emulsions ($\psi$ is low), where particles do not interact thermically [4]. 

$\lambda_{bed}/\lambda_{f}=\frac{1+2\frac{(1-\psi)\left(\frac{\lambda_p}{\lambda_f}-1\right)}{\left(\frac{\lambda_p}{\lambda_f}+2\right)}}{1-\frac{(1-\psi)\left(\frac{\lambda_p}{\lambda_f}-1\right)}{\left(\frac{\lambda_p}{\lambda_f}+2\right)}}=\frac{\lambda_p+2\lambda_f+2(1-\psi)(\lambda_p-\lambda_f)}{\lambda_p+2\lambda_f-(1-\psi)(\lambda_p-\lambda_f)}$

For beds, values are acceptable if these conditions are fulfilled:

1. Particles do not influence one another ==>> Diluted bed: $\psi<=0.40$
2. Moderate particle-to-fluid conductivity ratio: $\lambda_p/\lambda_f<20$.

Since $\lambda_{Ni,1000K}=72\frac{W}{mK}$ [5], $\lambda_{H_2,1000K}=0.40\frac{W}{mK}$ and $\lambda_{N_2,1000K}=0.061\frac{W}{mK}$, then $\rightarrow \lambda_{fuel}/\lambda_{Ni}\approx 0.0055 \text{ to } 0.00084$ and it is best to choose $\lambda_p=\lambda_{fuel}$ and $\lambda_f=\lambda_{Ni}$. In this case the porosity of the fuel channel is $\epsilon_{FC}=(1-\psi)$. The "diluted bed" condition is also fulfilled, since at low values of $\psi=\frac{V_f}{V_f+V_p}=\frac{V_{Ni}}{V_{Ni}+V_{fuel}}$, the value of $\epsilon_{FC}=(1-\psi)=\frac{V_{fuel}}{V_{Ni}+V_{fuel}}\approx 93\%$ is high.

$\lambda_{bed}/\lambda_{Ni}=\frac{\lambda_{fuel}+2\lambda_{Ni}+2\epsilon_{FC}(\lambda_{fuel}-\lambda_{Ni})}{\lambda_{fuel}+2\lambda_{Ni}-\epsilon_{FC}(\lambda_{fuel}-\lambda_{Ni})}$

```python
efc=linspace(0.6,1,30) # porosity of the fuel channel
k_Ni=72 # thermal conductivity of nickel at 1000K in W/m/K
fuel_labels=['H_2','CO_2']
k_fuel_values=[0.4,0.057] # thermal conductivity of fuel at 1000K in W/m/K

# plt.rcParams['axes.prop_cycle'] = cycler(
#     linestyle=['--','-.',':','-'])*cycler(color=['b', 'orange', 'g', 'r']) # set color and marker cycle

fig,ax=plt.subplots(2,2,sharex=True)
fig.set_size_inches(10*16/9,10)

for i in range(len(k_fuel_values)):
    k_fuel=k_fuel_values[i]
    k_Ni_foam=k_Ni*(k_fuel+2*k_Ni+2*efc*(k_fuel-k_Ni))/(k_fuel+2*k_Ni-efc*(k_fuel-k_Ni)) # Maxwell 1873 approach
    ax[0,0].plot(efc,k_Ni_foam,label='$'+r'\lambda_{fuel}=\lambda_{'+fuel_labels[i]+'}='+'{:g}~W/(m\cdot K)'.format(k_fuel)+'$')
# determine required k_fuel for a foam conductivity of 8.46W/m/K at porosity 0.867 by Maxwell approach
min_k_Ni_foam=min(k_Ni_foam)
k_fuel=72*((2-2*0.867-8.46/72*(2+0.867))/(-1-2*0.867+8.46/72*(1-0.867)))
k_Ni_foam=k_Ni*(k_fuel+2*k_Ni+2*0.867*(k_fuel-k_Ni))/(k_fuel+2*k_Ni-0.867*(k_fuel-k_Ni)) # Maxwell 1873 approach
ax[0,0].plot(0.867,k_Ni_foam,'o',color='black',fillstyle='full')
ax[0,0].plot([min(efc),0.867,0.867],[k_Ni_foam,k_Ni_foam,min_k_Ni_foam],'--',color='black')
ax[0,0].text(0.867*(1+.01),min_k_Ni_foam,r'$\epsilon_{FC}='+'{:g}'.format(0.867)+'$')
ax[0,0].text(0.867*(1+.01),k_Ni_foam,r'$\lambda_{Fuel}='+'{:g}'.format(k_fuel)+r'~\frac{W}{m\cdot K}'+'$')
ax[0,0].text(min(efc),k_Ni_foam,r'$\lambda_{Ni-foam}='+'{:g}'.format(k_Ni_foam)+r'~\frac{W}{m\cdot K}'+'$',horizontalalignment='left',verticalalignment='bottom')
k_fuel=k_fuel_values[0]
k_Ni_foam=k_Ni*(k_fuel+2*k_Ni+2*0.867*(k_fuel-k_Ni))/(k_fuel+2*k_Ni-0.867*(k_fuel-k_Ni)) # Maxwell 1873 approach
ax[0,0].plot([min(efc),0.867,0.867],[k_Ni_foam,k_Ni_foam,min_k_Ni_foam],'--')
ax[0,0].text(min(efc),k_Ni_foam,r'$\lambda_{Ni-foam}='+'{:g}'.format(k_Ni_foam)+r'~\frac{W}{m\cdot K}'+'$',horizontalalignment='left',verticalalignment='top')
#ax[0,0].set_xlabel(r'$\epsilon_{FC}$ / -')
ax[0,0].set_ylabel(r'$\lambda_{Ni-foam}$ / (W/m/K)')
ax[0,0].set_title(r'Maxwell approach $\lambda_{Ni-foam}=\lambda_{Ni}\frac{\lambda_{fuel}+2\lambda_{Ni}+2\epsilon_{FC}(\lambda_{fuel}-\lambda_{Ni})}{\lambda_{fuel}+2\lambda_{Ni}-\epsilon_{FC}(\lambda_{fuel}-\lambda_{Ni})}$',pad=20)
ax[0,0].legend()

for i in range(len(k_fuel_values)):
    k_fuel=k_fuel_values[i]
    k_Ni_foam=k_Ni/((1-efc)+efc*k_Ni/k_fuel)
    ax[1,0].plot(efc,k_Ni_foam,label='$'+fuel_labels[i]+'$')
ax[1,0].set_xlabel(r'$\epsilon_{FC}$ / -')
ax[1,0].set_ylabel(r'$\lambda_{Ni-foam}$ / (W/m/K)')
ax[1,0].set_title(r'Series approach $\lambda_{Ni-foam,I}=\left(\frac{1-\epsilon_{FC}}{\lambda_{Ni}}+\frac{\epsilon_{FC}}{\lambda_{fuel}}\right)^{-1}$')
ax[1,0].legend()

for i in range(len(k_fuel_values)):
    k_fuel=k_fuel_values[i]
    k_Ni_foam=(1-efc)*k_Ni+efc*k_fuel
    ax[1,1].plot(efc,k_Ni_foam,label='$'+fuel_labels[i]+'$')
ax[1,1].set_xlabel(r'$\epsilon_{FC}$ / -')
ax[1,1].set_ylabel(r'$\lambda_{Ni-foam}$ / (W/m/K)')
ax[1,1].set_title(r'Parallel approach $\lambda_{Ni-foam,II}=(1-\epsilon_{FC})\lambda_{Ni}+\epsilon_{FC}\lambda_{fuel}$')
ax[1,1].legend()

for a in [0.05,0.2,1-.35]: # series to parallel proportion
    for i in range(len(k_fuel_values)):
        k_fuel=k_fuel_values[i]
        k_Ni_foam_series=k_Ni/((1-efc)+efc*k_Ni/k_fuel)
        k_Ni_foam_parallel=(1-efc)*k_Ni+efc*k_fuel
        k_Ni_foam=(a/k_Ni_foam_series+(1-a)/k_Ni_foam_parallel)**-1
        ax[0,1].plot(efc,k_Ni_foam,label='a={:g}; '.format(a)+'$'+fuel_labels[i]+'$')
#ax[0,1].set_xlabel(r'$\epsilon_{FC}$ / -')
ax[0,1].set_ylabel(r'$\lambda_{Ni-foam}$ / (W/m/K)')
ax[0,1].set_title(r'$\lambda_{Ni-foam,combined}=\left(\frac{a}{\lambda_{Ni-foam,I}}+\frac{1-a}{\lambda_{Ni-foam,II}}\right)^{-1}$',pad=20)
ax[0,1].legend(loc='upper right')

plt.savefig('summary_series_parallel_approach.png',dpi=80)
```

## Series-parallel parameter $a$
Parameter $a$ adds a degree of freedom. Which value is optimal? In ref. [6], a best fit for experimental data of high porosity metal foams (Al-water, Al-air) was found to be $A=0.35$. Value $a=0.2$ in the Wärmeatlas [4] is arbitrary. 

Parameters $A$ from [6] and $a$ from [4] are analogous, but defined under different arrangements of the modelled mixed conductivity: $A$ is the proportion in series of an arrangement in **parallel** of $\lambda_{series}$ and $\lambda_{parallel}$, whearas $a$ is the proportion in series of an arrangement in **series** of the same $\lambda_{series}$ and $\lambda_{parallel}$. They are therefore related to each other as follows:

$\lambda_{series}=\left(\frac{1-\epsilon_{FC}}{\lambda_{Ni}}+\frac{\epsilon_{FC}}{\lambda_{fuel}}\right)^{-1}$

$\lambda_{parallel}=(1-\epsilon_{FC})\lambda_{Ni}+\epsilon_{FC}\lambda_{fuel}$

$A\cdot \lambda_{parallel}+(1-A)\cdot \lambda_{series}=\left(\frac{a}{\lambda_{series}}+\frac{1-a}{\lambda_{parallel}}\right)^{-1} \Leftrightarrow
a=\left(\frac{\lambda_{parallel}}{A\cdot \lambda_{parallel}+(1-A)\cdot \lambda_{series}}\right)\frac{\lambda_{series}}{\lambda_{parallel}-\lambda_{series}}$

```python
efc=0.867 # porosity of the fuel channel
A=0.35
print('porosity of the fuel channel: {:0.4g}'.format(efc))
print('Conductivity of Nickel at 1000K: {:0.4g} W/(m K)'.format(k_Ni))
print('Conductivity of fuel at 1000K and corresponding Ni-foam value assuming series/parallel approach with A={:0.4g}'.format(A))
for x in x_compositions: # loop each line of the compositions
    # calculate conductivity of gas mixture
    phi=1/sqrt(8)*(1+outer(M,1/M))**(-1/2)*(1+outer(viscosity,1/viscosity)**(1/2)*outer(1/M,M)**(1/4))**2
    viscosity_mix=sum(viscosity*x/phi.dot(x)) # semiempirical mixing rule
    conductivity_mix=sum(conductivity*x/phi.dot(x)) # semiempirical mixing rule
    heat_capacity_mix=heat_capacity.dot(x)
    density_mix=1e5/8.3145/T*M.dot(x)/1000
    
    k_fuel=conductivity_mix
    
    # calculate conductivity of foam by series and parallel approach
    k_Ni_foam_series=k_Ni/((1-efc)+efc*k_Ni/k_fuel)
    k_Ni_foam_parallel=(1-efc)*k_Ni+efc*k_fuel
    
    # calculate parameter a based on A
    a=(k_Ni_foam_parallel/(A*k_Ni_foam_parallel+(1-A)*k_Ni_foam_series)-1)*k_Ni_foam_series/(k_Ni_foam_parallel-k_Ni_foam_series)
    
    # calculate k_eff based on a value
    k_eff=(a/k_Ni_foam_series+(1-a)/k_Ni_foam_parallel)**-1
    
    gas_name=':'.join([labels[i] for i in range(len(labels)) if   x[i]!=0])+' '+':'.join(['{:d}'.format(int(round(  x[i]*100,0))) for i in range(len(labels)) if round(  x[i]*100,0)!=0])
    print(('{:30s}\tk_gas={:5.4g} W/(m K) \t-->>k_series={:5.4g} W/(m K), k_parallel={:5.4g} W/(m K), a={:5.4g}. k_eff={:5.4g} W/(m K)').format(gas_name,conductivity_mix,k_Ni_foam_series,k_Ni_foam_parallel,a,k_eff))
```
